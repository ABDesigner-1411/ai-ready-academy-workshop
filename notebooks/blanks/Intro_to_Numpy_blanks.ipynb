## Array Math
<b>Elementwise</b> array operations execute element by element operations on corresponding elements of arrays

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

# Do an Elementwise SUM of x and y using +
print(x + y)
```

```python
#Do the same elementwise SUM but use a method to do so:
print(np.add(x, y))
```

```python
#Do an Elementwise PRODUCT of x and y
print(x * y)
```

```python
# #Do the same elementwise PRODUCT but use a method to do so:
print(np.multiply(x, y))
```

<b>Matrix multiplication</b>: use the dot() function to compute inner products of arrays of the same size.<br/> Reminder: "*" is elementwise multiplication, not matrix multiplication.

```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

# dot product of x and y
print(x.dot(y))
# OR
print(np.dot(x,y))
```
